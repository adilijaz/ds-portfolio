### Categorical column cardinality
| Column | Unique values |
|---|---|
| order_status | 8 (delivered, shipped, canceled, processing, invoiced, unavailable, created, approved) |

| customer_id | 99,441 (all unique - one customer_id per order in this table) |

### Why the missing values exist
Missing values are not data entry errors - they reflect where an order stopped in its lifecycle:
- purchased -> approved -> handed to carrier -> delivered to customer
- An order missing `order_approved_at` was likely cancelled before approval
- An order missing `order_delivered_carrier_date` was approved but never shipped
- An order missing `order_delivered_customer_date` was shipped but not yet delivered (or lost)

### Verdict
Dataset is structurally clean (no duplicates, no random nulls) but requires datetime conversion on 5 columns before any time-based analysis (delivery time, monthly trends) is possible. Missing values should be handled by filtering on `order_status`, not by dropping rows blindly.
